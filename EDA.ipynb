```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
offers= pd.read_csv('total_offers.csv')
```

```python
offers.head()
```

```python
# Distribution of the "Price(EUR)"
plt.figure(figsize=(10, 6))
sns.histplot(data=offers, x='Price(EUR)', hue="Emission standard", palette="Set1")
plt.show()
```

As we can see, most of the prices are under 50000 Euro, and the majority of the offers are less than 10-year-old (Euro norm 6)

```python
# Let's first exam the ridiculously expansive offers
offers[offers['Price(EUR)']>=200000]
```

```python
# distribution of "Year"
sns.histplot(data=offers, x='Year') # majority of the offers are after 2015
```

```python
pd.set_option('display.max_rows', None)
```

```python
# how the price(mean) for each brand top 20
price_make = offers.groupby('Make')['Price(EUR)'].agg('mean').sort_values(ascending=False)[:20]
```

```python
# To see of the car brand affects the price
figure_size = (15, 6)
plt.figure(figsize=figure_size)
sns.barplot(data=offers, x=price_make.index, y=price_make.values)
for index, value in enumerate(price_make.values):
    plt.text(index, value, f"{round(value, 2)}", ha='center', va='bottom', rotation=90)
plt.xticks(rotation=45)
plt.title('Average price of Top20 brand')
plt.show()
```

The most expensive brands (Aston Martin, Lamborghini, Rolls Royce, Mclaren and Farrari) are having a huge different with the rest

```python
# Drill down to the model level
price_model_make = offers.groupby(['Make','Model'])['Price(EUR)'].agg('mean').sort_values(ascending=False)[:20].reset_index()
```

```python
figure_size = (15, 6)
plt.figure(figsize=figure_size)
ax= sns.barplot(data=price_model_make, x='Model', y='Price(EUR)')
# Annotate each bar with the corresponding Make value
for index, row in price_model_make.iterrows():
    ax.text(index, row['Price(EUR)'], row['Make'],
            color='black', ha='center', va='bottom', rotation=45)

plt.xticks(rotation=60)
plt.title('Average price of Top20 model')
plt.show()
```

Mercedes-Benz, Lamborghini, Rools Royce, Mclaren and Porsche have the most expensive models in the used car market

Check "Fuel type" , "Body type", "Transmission" and "Country" VS "Price"


```python
# check the distribution first
display(offers['Fuel type'].value_counts(ascending=False))
display(offers['Body type'].value_counts(ascending=False))
display(offers['Transmission'].value_counts(ascending=False))
display(offers['Country'].value_counts(ascending=False))
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=offers, x='Fuel type', y='Price(EUR)')
plt.title('Price VS Fuel type')
plt.show()
```

Most of the cars are driven by "Petrol" and "Diesel", and the price varies a lot in these two categories;
Hybrid and Electric have much less occupation in the used car market, and the price is relatively higer in terms of the IQR and median value than the other categories

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=offers, x='Body type', y='Price(EUR)')
plt.xticks(rotation=45)
plt.title('Price VS Body type')
plt.show()
```

It's obvious that Coupe has higher averge price than the others, and the price also varies from under 50,000 to over 300,000 Euro;
Convertible has highest potential to reach extremly high price range;
Van, Crossover, Hatchback, MPV and Mini are clustered in the lower price range under 50000 Euro

```python
plt.figure(figsize=(5,5))
sns.boxplot(data=offers, x='Transmission', y='Price(EUR)')
plt.title('Price VS Transmission')
plt.show()
```

Automatic cars in general have higher price than the Manual ones, and also have the opportunity to reach a very high price level

```python
plt.figure(figsize=(15,5))
sns.boxplot(data=offers, x='Country', y='Price(EUR)')
plt.xticks(rotation=45)
plt.title('Price VS Country')
plt.show()
```

The median price seems to be stabalized under 50,000 euro for all 16 countries. There are no big differences in terms of IQR

For the rest numerical columns, for instance "kW", "Mileage(km)", we use pairplot

```python
sns.pairplot(data=offers[['kW', 'Mileage(km)', 'Price(EUR)']])
```

The more powerful the car is, the higher price it can be labled;
Lower mileage tend to have higher price;



Now we check the heatmap for the correlation

```python
offers.info()
```

```python
numeric_columns = ['Mileage(km)', 'kW', 'Year', 'Emission standard', 'Price(EUR)']
# Calculate the correlation matrix
correlation_matrix = offers[numeric_columns].corr().abs()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title('Absolute Correlation Heatmap')
plt.show()
```

It's good that there's no collinearity between the numerical columns and again "kW" may be a good parameter for precting the price

```python
pd.set_option('display.max_rows', None)
```

```python
grouped_data = offers.groupby(['Make', 'Model'])['Trim'].apply(lambda x: list(set(x))).reset_index()
```

```python
grouped_data
```

```python
# Convert the grouped data into dictionary format
car_data = {}
for index, row in grouped_data.iterrows():
    make = row['Make']
    model = row['Model']
    trim = row['Trim']
    if make not in car_data:
        car_data[make] = {}
    car_data[make][model] = trim

print(car_data)

```

```python
import pickle
```

```python
# Save the dictionary to a pickle file
with open('car_data.pkl', 'wb') as f:
    pickle.dump(car_data, f)
```
