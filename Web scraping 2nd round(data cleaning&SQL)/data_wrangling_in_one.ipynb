```python
import pandas as pd
import re
```

```python
# List to store rearranged dataframes
rearranged_dfs = []
```

```python
column_names= ['Make', 'Model', 'Trim', 'Year', 'Mileage(km)', 'kW', 'Fuel type', 'Body type', 'Color', 'Transmission', 'Emission standard', 'Contact', 'Country', 'City', 'Price(EUR)']
        
```

```python
as_offers = pd.read_csv('cleaned_Austria_offers.csv')
```

```python
be_offers = pd.read_csv('cleaned_Belgium_offers.csv')
cr_offers = pd.read_csv('cleaned_Czech_offers.csv')
dm_offers= pd.read_csv('cleaned_Denmark_offers.csv')
fr_offers = pd.read_csv('cleaned_France_offers.csv')
gr_offers= pd.read_csv('cleaned_Germany_offers.csv')
hg_offers = pd.read_csv('cleaned_Hungary_offers.csv')
it_offers = pd.read_csv('cleaned_Italy_offers.csv')
nl_offers= pd.read_csv('cleaned_Netherlands_offers.csv')
nw_offers = pd.read_csv('cleaned_Norway_offers.csv')
pl_offers = pd.read_csv('cleaned_Poland_offers.csv')
pt_offers = pd.read_csv('cleaned_Portugal_offers.csv')
rm_offers = pd.read_csv('cleaned_Romania_offers.csv')
sp_offers = pd.read_csv('cleaned_Spain_offers.csv')
st_offers = pd.read_csv('cleaned_Switzerland_offers.csv')
sd_offers = pd.read_csv('cleaned_Sweden_offers.csv')
```

```python
# Iterate over each dataframe and rearrange columns
for df in [as_offers, be_offers, cr_offers, dm_offers, fr_offers, gr_offers, hg_offers, it_offers, nl_offers, nw_offers, pl_offers, pt_offers, rm_offers, sp_offers, st_offers, sd_offers]:
    rearranged_df = df[column_names]  # Rearrange columns according to the desired order
    rearranged_dfs.append(rearranged_df)
```

```python
# Concatenate the rearranged dataframes
concatenated_df = pd.concat(rearranged_dfs, ignore_index=True)
```
