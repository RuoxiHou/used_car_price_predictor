```python
from pycaret.regression import *
```

```python
regression_setup = setup(offers, target = 'Price(EUR)')
```

```python
best_model = compare_models()
```

```python
# Create the best model
final_model = create_model(best_model)
```

```python
# Tune the best model
tuned_model = tune_model(final_model)
# Evaluate the tuned model
evaluate_model(tuned_model)
```

```python
predict_model(tuned_model)
```

```python
# Save the final model to a file
save_model(final_model, 'best_model_raw_data')
```

We can see that the best trained models are Extra Trees Regressor and Extreme Gradient Boosting. In Sklearn, Random search will be used for hyperparameter tuning
