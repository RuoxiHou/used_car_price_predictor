```python
import pandas as pd
```

```python
offers_01 = pd.read_csv('europe_offers_01.csv')
offers_02 = pd.read_csv('europe_offers_02.csv')
offers_03 = pd.read_csv('europe_offers_03.csv')
```

```python
total_offers = pd.concat([offers_01, offers_02, offers_03], ignore_index=True)
total_offers
```

```python
total_offers.to_csv('total_offers.csv', index= False)
```
